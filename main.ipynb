```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import mplcyberpunk
import numpy as np

mean = 0    
std = 1 
xs = np.random.normal(mean, std, 50)
ys = np.random.normal(mean, std, 50)

figure(figsize=(6, 6), dpi=80)
plt.style.use("cyberpunk")
plt.axvline(x=0, color='violet')
plt.axhline(y=0, color='violet')
lim = max(xs + ys, key=lambda x: abs(x)) * 1.15
plt.ylim(-lim,lim)
plt.xlim(-lim,lim)
#plt.invert_xaxis()
#plt.invert_yaxis()
plt.scatter(xs, ys)
plt.text(lim*0.85, 0.25, "x", color='violet', fontsize=16)
plt.text(0.25, lim*0.85, "y", color='violet', fontsize=16)
plt.show()
```

```python
from math import e as e

lambda_ = 1
A_ = 0.01
x = np.linspace(-10, 10, 100)
f_x = lambda_ * e**(A_*x**2)

figure(figsize=(6, 6), dpi=80)
plt.plot(x, f_x)
ylim_ = max(abs(f_x))
plt.ylim(-ylim_, ylim_)
xlim_ = max(abs(x))
plt.xlim(-xlim_*10,xlim_*10)
plt.axvline(x=0, color='violet')
plt.axhline(y=0, color='violet')
plt.text(xlim_*10*0.85, 0.25, "x", color='violet', fontsize=16)
plt.text(xlim_, -ylim_*0.85, "y", color='violet', fontsize=16)
```

```python
from math import pi as pi

lambda_ = 3
x = np.linspace(-10, 10, 1000)
f_x = lambda_ * e**(-pi*lambda_**2*x**2)

figure(figsize=(6, 6), dpi=80)
plt.plot(x, f_x)
ylim_ = max(abs(f_x))
plt.ylim(-ylim_, ylim_*1.1)
xlim_ = max(abs(x))
#plt.xlim(-xlim_*10,xlim_*10)
plt.axvline(x=0, color='violet')
plt.axhline(y=0, color='violet')
#plt.text(xlim_*10*0.85, 0.25, "x", color='violet', fontsize=16)
#plt.text(xlim_, -ylim_*0.85, "y", color='violet', fontsize=16)
```

```python
from math import sqrt

def f_x(x, sigma_, mu_=0):
    return 1 / (sigma_ * sqrt(2 * pi)) * e**(-(x-mu_)**2/(2*sigma_**2))

sigma_1 = 0.8
x1 = np.linspace(-10, 10, 1000)
y1 = f_x(x1, sigma_1)

y2 = f_x(x1, sigma_1, 3)

y3 = f_x(x1, sigma_1, -3)

figure(figsize=(6, 6), dpi=80)
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x1, y3)
ylim_ = max(abs(y1))
plt.ylim(-ylim_, ylim_*1.1)
xlim_ = max(abs(x))
#plt.xlim(-xlim_*10,xlim_*10)
plt.axhline(y=0, color='violet')
plt.axvline(x=0, color='teal')
plt.axvline(x=3, color='violet')
plt.axvline(x=-3, color='yellow')

plt.text(0.3, -0.1, "mu=0", color='teal', fontsize=12)
plt.text(3.3, -0.1, "mu=3", color='violet', fontsize=12)
plt.text(-3.3, -0.1, "mu=-3", color='yellow', fontsize=12)
```
